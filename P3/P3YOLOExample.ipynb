## Model

*YOLO26n* jest najmniejszym modelem z rodziny YOLO26 (*You-Only-Look-Once*). Jest to model real-time detekcji obiektów.

Modele dostępne są we frameworku Ultralytics.

```python
from ultralytics import YOLO

model = YOLO("yolo26n.pt")
```

Model pretrenowany jest na zestawie COCO (Common Objects in Context) zawierającym 80 klas obiektów.

Atrybut 'names' przechowuje nazwy klas modelu we frameworku Ultralytics.

```python
model.names
```

## Przykładowe wnioskowanie

```python
import matplotlib.pyplot as plt

results = model("https://github.com/ultralytics/assets/releases/download/v0.0.0/bus.jpg")
# Results - lista wyników dla każdej przetworzonej ramki (w tym przypadku tylko jednej)


# Plot results image
im_bgr = results[0].plot()  # BGR-order numpy array

plt.imshow(im_bgr[:, :, ::-1])
plt.axis('off')
plt.show()

print(results)
```

```python
results = model("bus.jpg")

labels = results[0].names # Słownik z nazwami klas z datasetu

bboxes = results[0].boxes.xyxy # Tensor z współrzędnymi wykrytych obiektów w formacie [x_min, y_min, x_max, y_max]
confidences = results[0].boxes.conf # Tensor z wartościami pewności wykrycia obiektów (confidence scores)
classes = results[0].boxes.cls # Tensor z indeksami klas wykrytych obiektów (odwołuje się do listy klas w model.names)
```

```python
# Przykład "wyłuskania" informacji o wykrytych obiektach

labels = results[0].names # Pobranie nazw z datasetu (słownik)

results = model('https://github.com/ultralytics/assets/releases/download/v0.0.0/bus.jpg', verbose=False, conf=0.5) # Wykrywanie obiektów z ustawionym progiem pewności (confidence threshold)

bboxes = results[0].boxes.xyxy.detach().cpu().numpy() # Transformacja do NumPy array
confidences = results[0].boxes.conf.detach().cpu().numpy() # Transformacja do NumPy array
classes = results[0].boxes.cls.detach().cpu().numpy() # Transformacja do NumPy array

for i, (bbox, conf, cls) in enumerate(zip(bboxes, confidences, classes)):
    label = labels[int(cls)]  # Pobranie nazwy klasy na podstawie indeksu
    print(f"Obiekt {i+1}:")
    print(f"  Klasa: {label}")
    print(f"  Współrzędne: {bbox}")
    print(f"  Pewność wykrycia: {conf:.3f}")
```

```python
# Przykład zapisu do pliku TXT

with open("results.txt", "a") as f:
    for i, (bbox, conf, cls) in enumerate(zip(bboxes, confidences, classes)):        
        f.write(f"class: {labels[int(cls)]}, bbox: {bbox.tolist()}, confidence: {float(conf):.3f}\n")
```
