```python
# Przed uruchomieniem kodu należy przesłać foldery data_normal i data_abnormal do COLAB

import pandas as pd
import glob
import matplotlib.pyplot as plt
import numpy as np

files_normal = glob.glob("data_normal/*.csv")
list_dataframes_normal = [pd.read_csv(f, sep=";") for f in files_normal]

files_abnormal = glob.glob("data_abnormal/*.csv")
list_dataframes_abnormal = [pd.read_csv(f, sep=";") for f in files_abnormal]
```

### Przykładowa analiza sygnałów z SenseHat

Mamy dwa sygnały (przebiegi czasowe):
* pitch,
* roll,

opisują one kątowe położenie naszego urządzenia.
Wszystkie przebiegi mają długość 100 próbek co odpowiada czasu około 5 sekund.

Sygnały zostały zebrane w dwóch grupach:
* normal - zapis podczas normalnej pracy urządzenia,
* abnormal - zapis podczas poruszania urządzeniem.

Na poniższych wykresach zawarto następujące dane:
* Surowy sygnał
* Sygnał po filtracji (okno ruchome o N=11)
* Pochodna (diff) sygnału surowego
* Pochodna sygnału odfiltrowanego
* Średnie odchylenie standardowe dla zapisu

```python
fig, axes = plt.subplots(nrows=len(list_dataframes_normal), ncols=2, figsize=(10, 3*len(list_dataframes_normal)))

for i, df in enumerate(list_dataframes_normal):
    t = df["timestamp"]
    pitch = df["pitch"].to_numpy()
    roll = df["roll"].to_numpy()

    pitch_filtered = np.convolve(pitch, np.ones(11)/11, mode='valid')
    roll_filtered = np.convolve(roll, np.ones(11)/11, mode='valid')
    
    diff_pitch = np.diff(pitch, prepend=pitch[0])
    diff_roll = np.diff(roll, prepend=roll[0])

    diff_pitch_filtered = np.diff(pitch_filtered, prepend=pitch_filtered[0])
    diff_roll_filtered = np.diff(roll_filtered, prepend=roll_filtered[0])

    mean_std_pitch = np.mean(np.std(pitch))
    mean_std_roll = np.mean(np.std(roll))
    mean_std_pitch_filtered = np.mean(np.std(pitch_filtered))
    mean_std_roll_filtered = np.mean(np.std(roll_filtered))

    axes[i][0].plot(t, pitch, color=f"C0", linestyle='-', alpha=0.75, linewidth=1)
    axes[i][0].plot(t[5:-5], pitch_filtered, color=f"C0", linestyle='--', linewidth=2)
    axes[i][0].plot(t, diff_pitch, color=f"C1", linestyle='-', alpha=0.75, linewidth=1)
    axes[i][0].plot(t[5:-5], diff_pitch_filtered, color=f"C1", linestyle='--', linewidth=2)
    axes[i][0].text(0.05, 0.9, f"Mean std.: {mean_std_pitch:.4f} | {mean_std_pitch_filtered:.4f}", transform=axes[i][0].transAxes, fontsize=10, color='r', ha='left')

    axes[i][0].set_xlabel("Timestamp")
    axes[i][0].set_ylabel("Pitch")

    axes[i][0].set_title(f"Data from {files_normal[i]}")
    axes[i][0].grid(linestyle='--', alpha=0.5)
    axes[i][0].legend(["Pitch", "Pitch Filtered", "Diff Pitch", "Diff Pitch Filtered"], loc="upper right", fontsize=7)


    axes[i][1].plot(t, roll, color=f"C0", linestyle='-', alpha=0.75, linewidth=1)
    axes[i][1].plot(t[5:-5], roll_filtered, color=f"C0", linestyle='--', linewidth=2)
    axes[i][1].plot(t, diff_roll, color=f"C1", linestyle='-', alpha=0.75, linewidth=1)
    axes[i][1].plot(t[5:-5], diff_roll_filtered, color=f"C1", linestyle='--', linewidth=2)
    axes[i][1].text(0.05, 0.9, f"Mean std.: {mean_std_roll:.4f} | {mean_std_roll_filtered:.4f}", transform=axes[i][1].transAxes, fontsize=10, color='r', ha='left')
   
    axes[i][1].set_xlabel("Timestamp")
    axes[i][1].set_ylabel("Roll")

    axes[i][1].set_title(f"Data from {files_normal[i]}")
    axes[i][1].grid(linestyle='--', alpha=0.5)
    axes[i][0].legend(["Pitch", "Pitch Filtered", "Diff Pitch", "Diff Pitch Filtered"], loc="upper right", fontsize=7)

plt.tight_layout()
plt.show()
```
